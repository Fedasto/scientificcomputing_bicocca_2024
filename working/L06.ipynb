## Q5: Latexdiff

Latexdiff is a popular tool to mark differences between two latex files.

*latexdiff old.tex new.tex > diff.tex*

This will produce a tex file with highlights in blue and red, marking all the words that have been changed between the two versions. Incredibly useful! (It's the same tool that many journal editors use when they send you the proofs of the paper after acceptance).

Put a tex file under version control, edit it, add/commit/push it, then edit it again, and add/commit/push it again. Generate the latexdiff between those two commits.

```python
from IPython.display import display, Math, Markdown

# The LaTeX code
latex_code = r"""
\medskip For example, see Equation \ref{eq:1} and Equation \ref{eq:2} below.
\begin{equation} \label{eq:1}
     a^2 + b^2 \DIFaddbegin \DIFadd{+ d^2 }\DIFaddend = c^2
\end{equation}
\begin{equation} \label{eq:2}
\begin{split}
A & = \frac{\pi r^2}{2} \\
  & = \frac{1}{2} \pi r^2 \DIFaddbegin \DIFadd{*\int{0}{\inf}{2x}
\end{split}
\end{equation}
"""

# Display the LaTeX
display(Markdown(latex_code))
```
