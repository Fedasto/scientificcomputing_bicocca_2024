# Q2: Planetary positions

Plot as points, the periods vs. distances for each planet on a log-log plot.

Write the name of the planet next to the point for that planet on the plot

```python
a = np.array([0.39, 0.72, 1.00, 1.52, 5.20, 9.54, 19.22, 30.06, 39.48]) # semimajor axis
P = np.array([0.24, 0.62, 1.00, 1.88, 11.86, 29.46, 84.01, 164.8, 248.09]) # periods
names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", 
         "Uranus", "Neptune", "Pluto"]

fig, ax = plt.subplots(figsize=(8,8))



for i in range(len(names)):
    ax.scatter(a[i],P[i], marker='o', alpha=0.8)
    ax.text(a[i],P[i],names[i],transform=ax.transData)

# ax.set_xscale('log')
# ax.set_yscale('log')
ax.loglog()
ax.set_title('Period vs Semimajor-ax')
ax.set_xlabel(r"$semimajoraxis$")
ax.set_ylabel(r'$period$')
plt.axis('equal')

```

# Q5: Subplots

matplotlib has a number of ways to create multiple axes in a figure -- look at plt.subplots() (http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.subplot)

Create an x array using NumPy with a number of points, spanning from .

Create 3 axes vertically, and do the following:

Define a new numpy array f initialized to a function of your choice.
Plot f in the top axes
Compute a numerical derivative of f,
 
and plot this in the middle axes
Do this again, this time on 
 to compute the second derivative and plot that in the bottom axes

```python
t = np.linspace(0,2*np.pi,360)

def f(x):
    return np.sin(x)

def f_prime(y,x):
    derivative=[]
    for i in range(len(x)-1):
         derivative.append((y[i+1] - y[i]) / (x[i+1] - x[i]))
    return derivative

func = f(t)
der_func = f_prime(func,t)
sec_der_func = f_prime(der_func, t[:-2])

fig,ax = plt.subplots(3,1,figsize=(12,8), sharex=True)

ax[0].plot(func, color='blue')
ax[0].annotate(r"$f(x)=\sin(x)$",(20,-0.5))
ax[0].set_xticks(np.linspace(0,360,5),['0',r'$\frac{\pi}{2}$', r'$\pi$',r'$\frac{3\pi}{2}$', r'$2\pi$'])
ax[1].plot(der_func, color='orange')
ax[1].annotate(r"$f'(x)=\cos(x)$",(20,-0.6))
ax[2].plot(sec_der_func, color='green')
ax[2].annotate(r"$f''(x)=-\sin(x)$",(20,0.5))

plt.subplots_adjust(hspace=0.0)
#plt.tight_layout()
```
